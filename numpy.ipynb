```python
memory = np.array([0,1,1,0,0,1,1,1,0,1,0])

diff = np.diff(memory)       #The diff() function returns values of -1 or 1 for those indices, respectively.
starts = np.nonzero(diff == 1)[0]+1
ends = np.nonzero(diff == -1)[0]
sizes = ends -starts

indices = np.argsort(sizes)

for i in range(len(sizes)):
    print((starts[indices[i]],ends[indices[i]],sizes[indices[i]]))
```

```python
bookshelf = ['A','B','C','D','J','Q','W']
def binary_search(title):
        lower = 0
        upper = len(bookshelf)
        idx = int(np.mean([lower,upper]))
        
        while not bookshelf[idx] == title:
            if lower == idx or upper == idx:
                idx = -1
                break
            elif title < bookshelf[idx]:
                upper = idx
            else:
                lower = idx
                idx = int(np.mean([lower,upper]))
        return idx
print(binary_search('Q'))
print(binary_search('Y'))
```
