```python
def linear_search(test,key):
    for i in range(0,len(test)):
        if(test[i]==key):
            return i
    return -1
list1 = ["Python","Flask","Django","Seaborn","Matplotlib","Numpy","Pandas"]
key="Django"
result =linear_search(list1,key)
if result !=-1:
    print("The index of key is at {0}.".format(result))
else:
    print("The key was not found.")
```

```python
def binary_search(list1,src):
    low =0
    high=len(list1)-1
    mid=0
    while low<=high:
        mid = (low+high)//2
        if list1[mid]<src:
            low = mid-1
        elif list1[mid]>src:
            high = mid-1
        else:
            return mid
    return -1
list1 = [2,3,5,7,11,13,15,17,19,21,23,25]
key = 17
result = binary_search(list1,key)
if result == -1:
    print("Number is not found in list.")
else:
    print("Number is present at index.",result)
```

```python
def insertion_sort(list1):
    for i in range(1,len(list1)):
        key = list1[i]
        j=i-1
        while j>=0 and key<list1[j]:
            list1[j+1]=list1[j]
            j-=1
        list1[j+1]=key
    return list1
list1 = [11,5,12,9,17,7]
print(insertion_sort(list1))
```

```python
from heapq import heappop, heappush
def heap_sort(list1):
    heap=[]
    for i in list1:
        heappush(heap,i)
    sort = []
    while heap:
        sort.append(heappop(heap))
    return sort
list1=[2,3,11,22,32,54,39]
print(heap_sort(list1))
```
