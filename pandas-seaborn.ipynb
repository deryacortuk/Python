```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [2, 4, 1]
fig = plt.figure(figsize=(8,5))
axes =fig.add_axes([0,0,1,1])
axes.set_title("İki farklı değer çizimi")
axes.set_xlabel("x-ekseni")
axes.set_ylabel("y-ekseni")
axes.plot(x, y, linestyle='-')

plt.grid()
plt.show()
```

```python
x1 = [10,20,30]
y1 = [20,40,10]
y2 = [40,10,30]
fig = plt.figure(figsize=(8,5))
axes =fig.add_axes([0,0,1,1])
axes.set_title("İki farklı değer çizimi")
axes.set_xlabel("x-ekseni")
axes.set_ylabel("y-ekseni")
axes.plot(x1,y1,color="blue",label="A-mavi",lw=3,linestyle="--" )
axes.plot(x1,y2,color="red",label="B-kırmızı",lw=7,linestyle="--" )
axes.grid()
axes.legend()
```

```python
import numpy as np


x = np.arange(0,4*np.pi,0.01)
y = 3*np.sin(x) - np.cos(x)

fig,axes = plt.subplots(figsize=(7,7))

axes.plot(x, y,color ="blue",lw=3)
axes.set(xlabel='radians', ylabel='3*Sin(x) - cos(x)',title='Trigonometry')
axes.grid()

plt.show()
```

```python
fig =plt.figure(figsize=(9,9))
axes =fig.add_axes([0,0,1,1])
x = np.arange(0,4*np.pi,0.01)
y = 3*np.sin(x) - np.cos(x)
z = 2*np.cos(x)*np.sin(x)
axes.plot(x,y,color="purple",label="3*sin(x)-cos(x)")
axes.plot(x,z,color="green",label="2*cos(x)*sin(x)")
axes.grid()
axes.legend()
```
